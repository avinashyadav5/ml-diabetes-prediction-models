ROC CURVE AND AREA UNDER CURVE

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
train_x_norm=min_max_scaler.fit_transform(train_x)
test_x_norm=min_max_scaler.fit_transform(test_x)
```

```python
train_x_norm[0,0]
```

```python
train_x.iloc[0,0]
```

```python
xgbm=OneVsRestClassifier(xgb.XGBClassifier(max_depth=20))
xgbm.fit(train_x_norm,train_y)
y_pred = xgbm.predict(train_x_norm)
pred_prob=xgbm.predict_proba(test_x_norm)
```

```python
test_y_binarized=label_binarize(test_y,classes=np.unique(test_y))
fpr = {}
tpr = {}
thresh={}
roc_auc=dict()
n_class=classes.shape[0]
for i in range(n_class):
 fpr[i], tpr[i], thresh[i]= roc_curve(test_y_binarized[:, i], pred_prob[:, i])
 roc_auc[i] = auc(fpr[i], tpr[i])
 
for i in range(n_class):
 plt.plot(fpr[i], tpr[i], linestyle='--',
 label='%s vs Rest (AUC=%0.2f)'%(classes[i],roc_auc[i]))
plt.plot([0,1],[0,1],'b--')
plt.xlim([0,1])
plt.ylim([0,1.05])
plt.title('Multiclass ROC curve')
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.legend(loc='lower right')
plt.show()
```
